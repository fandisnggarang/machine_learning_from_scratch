```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 

from KMeans_Clustering import KMeans_Clustering
from sklearn.metrics import silhouette_score, adjusted_rand_score, normalized_mutual_info_score

np.random.seed(42)
from sklearn.datasets import make_blobs

X, y = make_blobs(
    centers=3, n_samples=500, n_features=2, shuffle=True, random_state=40
)
print(f'Shape of X        : {X.shape}')

clusters = len(np.unique(y))
print(f'Number of clusters: {clusters}')
```

```python
k = KMeans_Clustering(k=clusters, num_iters=150, plot_steps=True, tol=1e-4, random_state=42, distance_measurement='manhattan')
y_pred = k.predict(X)

df = pd.DataFrame(data = X, columns=['Feature1', 'Feature2'])
df['Y']     = y
df['label'] = k.labels
print(df) 

print("\nSilhouette Score:")
print(silhouette_score(X, k.labels))

print("\nARI:")
print(adjusted_rand_score(y, k.labels))

print("\nNMI:")
print(normalized_mutual_info_score(y, k.labels))
```
